```python
# Funkcja GLA

def Kraków_GLA(p_działki):
    min_liczba_mp=30
    pow_mp=15
    wsk_wspolne=0.3
    maks_GLA=(5000/(min_liczba_mp*pow_mp))*1000
    
    wsk_int=((maks_GLA/(1-wsk_wspolne))/p_działki)
    
    if wsk_int>=1.9 and wsk_int<=2.5:
        calc_GLA=maks_GLA
    elif wsk_int<1.9:
        calc_GLA=0
    elif wsk_int>2.5:
        calc_GLA=p_działki*2.5*(1-wsk_wspolne)
    
    return calc_GLA
```

```python
def Wycena_DCF(gla,pc,wsk_pust,czynsz_m,oe_m,op_m,proj_pnb,capex,nadzor,rezerwa,okres,fin_obce,fin_st,komerc,st_dysk):
    
    #---------Dochód-------------------------------------------------
    
    p_czynsz_efektywny_r=gla*czynsz_m*12*(1-wsk_pust/100)
    p_opłata_eksploatacyjna_r=gla*(1-wsk_pust/100)*oe_m*12
    p_opłata_garaż_r=((gla/1000)*30).astype(int)*(1-wsk_pust/100)*op_m*12
    k_eksploatacyjne_r=-gla*oe_m*12-((gla/1000)*30).astype(int)*op_m*12
    
    #---------CAPEX--------------------------------------------------
    
    i_projekt=-gla*proj_pnb
    i_capex=-gla*capex
    i_nadzor=nadzor*i_capex/100
    i_rezerwa=rezerwa*i_capex/100
    
    i_capex_all=i_projekt+i_capex+i_nadzor+i_rezerwa
    
    i_fin=0
    for x in range(1,int(okres)+1):
        i_fin=i_fin+((i_capex_all*fin_obce/100)/int(okres))*x*fin_st/1200
    
    i_komerc=-p_czynsz_efektywny_r/12*komerc
    
    #------Wycena----------------------------------------------------
    
    w_DON=p_czynsz_efektywny_r+p_opłata_eksploatacyjna_r+p_opłata_garaż_r+k_eksploatacyjne_r
    w_REA=w_DON/(st_dysk/100)
    
    w_DCF=w_REA+i_capex_all+i_fin+i_komerc
    
    return [w_DCF,w_REA,w_DON,i_capex_all+i_fin+i_komerc,p_czynsz_efektywny_r,p_opłata_eksploatacyjna_r,p_opłata_garaż_r,k_eksploatacyjne_r,i_projekt,i_capex,i_nadzor,i_rezerwa,i_fin,i_komerc]
    
```
